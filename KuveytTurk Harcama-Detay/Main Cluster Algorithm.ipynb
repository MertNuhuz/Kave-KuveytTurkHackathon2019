```python
#Kütüphanelerin içeri alınması
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None) 
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_colwidth', -1)
```

```python
#Datanın içeri alınması
data = pd.read_excel("maindata.xlsx")
data.head()
```

```python
# %99 üstü Boş olan değişkenlerin düşürülmesi
def NanCleaner(columnlist):
    drop = []
    for i in columnlist:
        if data[i].isnull().sum() >= len(data[i]) * 0.99:
            drop.append(i)
        else:
            continue
    print("NaN Olanlar",drop)
    return drop

data.drop(NanCleaner(data.columns), axis=1, inplace=True)
```

```python
#Temiz verinin çekilmesi, bu yukarıdaki işlemler + manuel işlemler ile temizlendi.
newdata = pd.read_excel("NaNClean.xlsx")
CustomerId = newdata.CustomerId
productkodu = newdata.kaProductKodu
newdata.drop("CustomerId",axis=1,inplace=True)
newdata = newdata.drop_duplicates()
newdata.head()
```

```python
newdata.info()
```

```python
#Kategorik değişkenlerin handle edilmesi
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
objects = ["kaProductKodu"]
for i in objects:  
    le.fit(newdata[i])
    newdata[i] = le.transform(newdata[i])
    
filename = 'labelencoder.sav'
pickle.dump(le, open(filename, 'wb'))

kategori = ["kaEkstreStatu","kaEMailOnlyFlag","EkstreTalimatEmailEv","EkstreTalimatGSMSMS"]
for i in kategori:
    newdata[i] = newdata[i].astype("category").cat.codes
```

```python
#Outlier Detection

from sklearn.ensemble import IsolationForest
isolation_forest = IsolationForest(n_estimators=100, contamination = 0.01)
isolation_forest.fit(newdata)

outlier = isolation_forest.predict(newdata)
outliers = np.where(outlier==-1)[0]

newdata.drop(outliers, axis=0, inplace=True)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
%matplotlib inline
from sklearn import datasets

km = KMeans(n_clusters=4)

X = newdata[["kaProductKodu","KartSubeKodu","kaEkstreStatu"]]
km.fit(X)
km.predict(X)
labels = km.labels_
```

```python
newdata["Grup"] = labels
```
